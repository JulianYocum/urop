```python
def get_exposure(hist, bins, phi, theta):
    xedges = np.linspace(0,np.pi/2,bins[0]+1)
    yedges = np.linspace(0,1, bins[1]+1)
    
    i_index = 0
    for i in range(bins[0]):
        if phi < xedges[i+1]:
            i_index = i
            break
            
    j_index = 0
    for j in range(bins[1]):
        if theta < yedges[j+1]:
            j_index = j
            break
            
    #print(i_index, j_index)
            
    return hist[i_index][j_index]
```
