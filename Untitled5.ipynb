```python
average_pulses = []
outliers = []
for c in range(1, 990):
    pulses = 0
    channelevents = noisydata.get_eventdf()[noisydata.get_eventdf()['Channel'] == c]
    for i in channelevents['NumPulses']:
        pulses += i
    
    if len(channelevents) > 0:
        average_pulse = pulses/len(channelevents)
        average_pulses.append(average_pulse)
        
        if average_pulse >= 1.5:
            outliers.append(c)
            
    else:
        average_pulses.append(0)
        
    
```
