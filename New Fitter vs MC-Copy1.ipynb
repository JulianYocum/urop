```python
from Helper import *
import copy

clean_test = copy.deepcopy(test)
clean_test.filter_fit(100,5)
# clean_test.clusterdf = clean_test.clusterdf[clean_test.clusterdf['ExtraCh'] == 0]
# clean_test.eventdf = clean_test.eventdf[clean_test.eventdf['Cluster'].isin(clean_test.clusterdf['Cluster'].values)]

clean_mc = copy.deepcopy(mc)
clean_mc.filter_fit(100,5)
# clean_mc.clusterdf = clean_mc.clusterdf[clean_mc.clusterdf['Cluster'].isin(clean_test.clusterdf['Cluster'].values)]
# clean_mc.eventdf = clean_mc.eventdf[clean_mc.eventdf['Cluster'].isin(clean_test.clusterdf['Cluster'].values)]


dEdx_pdf = make_pdf('./data/pdf/dEdx/bins_dEdx_muon.csv', './data/pdf/dEdx/values_dEdx_muon.csv', domain_range=[0,190])
clusters = np.array(mc.get_clusterdf()['Cluster'])

log_likelihoods_mc = [np.log(dEdx_pdf(np.array(clean_mc.get_cluster(c)['dEdx'])/100)).sum() for c in clusters]
log_likelihoods_test = [np.log(dEdx_pdf(np.array(clean_test.get_cluster(c)['dEdx'])/100)).sum() for c in clusters]


# min_path_test = test.get_eventdf()[test.get_eventdf()['PathLength'] > 30]
# min_path_mc = mc.get_eventdf()[test.get_eventdf()['PathLength'] > 30]

plt.figure(figsize=(12,8))

plt.subplot(2,3,1)
plt.title("dEdx")
#plt.scatter(np.array(mc.get_eventdf()['dEdx'])/100, np.array(test.get_eventdf()['dEdx'])/100, s=5)
plt.scatter(clean_test.get_eventdf()['dEdx']/100, clean_mc.get_eventdf()['dEdx']/100, s=5)
plt.xlabel("MeV/cm (reconstructed)")
plt.ylabel("MeV/cm (generated)")
plt.xlim(4,30)
plt.ylim(4,30)
plt.xscale('log')
plt.yscale('log')

plt.subplot(2,3,2)
plt.title("dEdx log-likelihoods")
plt.scatter(log_likelihoods_mc, log_likelihoods_test, s=5)
plt.xlim(-100,0)
plt.ylim(-100,0)
plt.xlabel("log(likelihood) (generated)")
plt.ylabel("log(likelihood) (reconstructed)")

plt.subplot(2,3,3)
plt.title("Pathlengths")
plt.scatter( np.array(clean_mc.get_eventdf()['PathLength']), np.array(clean_test.get_eventdf()['PathLength']), s=5)
plt.ylabel("mm (reconstructed)")
plt.xlabel("mm (generated)")
plt.xlim(0,80)
plt.ylim(0,80)

plt.subplot(2,3,4)
plt.title("Zenith")
plt.scatter(np.array(clean_mc.get_clusterdf()['Zenith']) * 180 / np.pi, np.array(clean_test.get_clusterdf()['Zenith']) * 180 / np.pi, s=5)
plt.xlabel("degrees (reconstructed)")
plt.ylabel("degrees (generated)")


plt.subplot(2,3,5)
plt.title("Azimuth")
plt.scatter(np.array(clean_mc.get_clusterdf()['Azimuth']) * 180 / np.pi, np.array(clean_test.get_clusterdf()['Azimuth']) * 180 / np.pi, s=5)
plt.xlabel("degrees (reconstructed)")
plt.ylabel("degrees (generated)")

plt.tight_layout()
```
