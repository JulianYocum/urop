```python

time_thresholds = [.1, .5, 1, 2]
events = 5
channels = 5
NRMSE = 2.5
#var_timedata = []

for t in time_thresholds:
    data = DataAnalysis().filter_noisy().filter_baseline().arrange_clusters(events, t).make_clusterdf().filter_clusterdf(NRMSE, channels)
    print("Cluster rate: " + str(data.get_clusterrate()))
    
    file_var = "t=" + str(t)
    data.save_dfs("data/variable/eventdf_" + file_var, "data/variable/clusterdf_" + file_var)
    
    #var_timedata.append(data.get_clusterdf()['Channels'])
```
